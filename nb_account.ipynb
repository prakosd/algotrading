```python
from datetime import datetime
from algotrading.backtesting.components.account.api import EIAccount
from algotrading.backtesting.components import ComponentFactory

account: EIAccount = ComponentFactory.create_account(datetime.now(), 100000)
print(account)
print(f"margin: {account.margin}")
account.deposit(datetime.now(), 2500, "2nd Deposit")
account.withdraw(datetime.now(), 500, "1st Withdrawal")
account.deposit(datetime.now(), 5000, "3rd Deposit")
account.withdraw(datetime.now(), 1000, "2nd Withdrawal")
account.reset_balance(datetime.now(), "Let's Go")
account.margin_lock(datetime.now(), 2000, "Pinjem dulu ya")
print(f"margin: {account.margin}")
account.close_trade(datetime.now(), 2000, 10, "Horee")
account.margin_release(datetime.now(), 200, "Aku balikin nih")
account.margin_lock(datetime.now(), 200, "Pinjem dulu ya")
account.margin_lock(datetime.now(), 150, "Pinjem dulu ya")
print(f"margin: {account.margin}")
account.close_trade(datetime.now(), 100, -60, "Damn")
account.margin_release(datetime.now(), 200, "Aku balikin nih")
print(f"{account.BASE_CURRENCY} {account.actual_balance} (initial: {account.BASE_CURRENCY} {account.initial_balance})")
print(EIAccount.BASE_CURRENCY)
print(EIAccount.LEVERAGE)
print(EIAccount.UNIT_SIZE)
account.get_ledger()
# account.clear_ledger()
```
