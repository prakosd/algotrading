```python
from datetime import datetime
from algotrading.backtesting.components.account import Account
from algotrading.backtesting.components import ComponentFactory

account: Account = ComponentFactory.create_account(datetime.now(), 100000)
print(account)
print(f"margin: {account.margin}")
account.deposit(datetime.now(), 2500, "2nd Deposit")
account.withdraw(datetime.now(), 500, "1st Withdrawal")
account.deposit(datetime.now(), 5000, "3rd Deposit")
account.withdraw(datetime.now(), 1000, "2nd Withdrawal")
account.reset_balance(datetime.now(), "Let's Go")
account.margin_lock(datetime.now(), 2000, "I owe you this")
print(f"margin: {account.margin}")
account.close_trade(datetime.now(), 2000, 10, "closing my trade")
account.margin_release(datetime.now(), 200, "I return the margin")
account.margin_lock(datetime.now(), 200, "May I borrow this")
account.margin_lock(datetime.now(), 150, "Lend me the money")
print(f"margin: {account.margin}")
account.close_trade(datetime.now(), 100, -60, "I'm closing my trade")
account.margin_release(datetime.now(), 200, "I pay it back")
print(f"{account.DEPOSIT_CURRENCY} {account.actual_balance} (initial: {account.DEPOSIT_CURRENCY} {account.initial_balance})")
print(Account.DEPOSIT_CURRENCY)
print(Account.LEVERAGE)
print(Account.UNIT_SIZE)
account.get_ledger()
# account.clear_ledger()
```
