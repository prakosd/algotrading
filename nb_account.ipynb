```python
from datetime import datetime
from algotrading.backtesting.components.account import Account

dt_test = datetime.strptime("2024-06-30", "%Y-%m-%d")
account = Account(dt_test, 100000)
print(account)
print(f"margin: {account.margin}")
account.deposit(dt_test, 2500, "2nd Deposit")
account.withdraw(dt_test, 500, "1st Withdrawal")
account.deposit(dt_test, 5000, "3rd Deposit")
account.withdraw(dt_test, 1000, "2nd Withdrawal")
account.reset_balance(dt_test, "Let's Go")
account.margin_lock(dt_test, 2000, "I owe you this")
print(f"margin: {account.margin}")
account.close_trade(dt_test, 2000, 10, "closing my trade")
account.margin_release(dt_test, 200, "I return the margin")
account.margin_lock(dt_test, 200, "May I borrow this")
account.margin_lock(dt_test, 150, "Lend me the money")
print(f"margin: {account.margin}")
account.close_trade(dt_test, 100, -60, "I'm closing my trade")
account.margin_release(dt_test, 200, "I pay it back")
print(f"{account.DEPOSIT_CURRENCY} {account.actual_balance} (initial: {account.DEPOSIT_CURRENCY} {account.initial_balance})")
print(Account.DEPOSIT_CURRENCY)
print(Account.LEVERAGE)
print(Account.UNIT_SIZE)
account.get_ledger()
# account.clear_ledger()
```
