```python
from algotrading.common.trade import Timeframe
from algotrading.common.asset import AssetPairCode as Symbol
from algotrading.providers import ProviderFactory

provider = ProviderFactory.create_provider_by_name(
    name = ProviderFactory.OANDA,
    symbol = Symbol.EUR_USD,
    start = "2023-01-01",
    end = "2023-06-30",
    timeframe = Timeframe.HOUR_1
)

ticker = provider.get_ticker()
print(ticker)
print(ticker.symbol.value)
print(ticker.timeframe)
print(ticker.get_digit())
tick = ticker.get_data(-1)
print(tick)
tick = ticker.get_data(datetime = "2023-01-02 14:00:00+00:00")
print(tick)
data = ticker.get_data()
print(data.index[1])
print(ticker.get_timezone())
```

```python
ticker.analyze()
```

```python
ticker.plot()
```

```python
ticker.resample(Timeframe.DAY_1)
print(ticker.timeframe)
ticker.plot()
```
