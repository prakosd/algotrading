```python
from algotrading import OandaProvider, Granularity

symbol    = "AUD_JPY"
start     = "2018-01-01"
end       = "2023-06-30"
timeframe = Granularity.DAY_1

provider = OandaProvider(symbol, start, end, timeframe)
print(provider)
print(provider.get_filename())
data = provider.get_response(True)
data
```

```python
from algotrading import ProviderFactory, EIProvider, Granularity

symbol    = "AUD_CAD"
start     = "2018-01-01"
end       = "2023-06-30"
timeframe = Granularity.DAY_1

provider: EIProvider = ProviderFactory.create_provider_by_name(
    name=EIProvider.PROVIDER_OANDA,
    symbol=symbol,
    start=start,
    end=end,
    granularity=timeframe
)
print(provider)
print(provider.get_filename())
data = provider.get_response()
data
```
